```python
import pandas as pd
weather = pd.read_csv('assets/weather.csv')
```

```python
def eda(dataframe):
    print("missing values \n", dataframe.isnull().sum(), '\n')
    print("dataframe index \n", dataframe.index, '\n')
    print("dataframe types \n", dataframe.dtypes, '\n')
    print("dataframe shape \n", dataframe.shape, '\n')
    print("dataframe describe \n", dataframe.describe(include='all'), '\n')
eda(weather)
```

```python
train_modified = train.groupby('Date')
```

```python
print(list(train_modified)[1][1]['NumMosquitos'])
```

```python
mosquitos_dict = {}
for tupl in list(train_modified):
    mosquitos_dict[tupl[0]] = tupl[1]['NumMosquitos'].sum()
print(mosquitos_dict)
```

```python
mosquitos_frame = pd.DataFrame([mosquitos_dict.values(), mosquitos_dict.keys()])
mosquitos_frame = mosquitos_frame.transpose()
```

```python
mosquitos_frame.columns = ['NumMosquitos', 'Date']
```

```python
mosquitos_frame = mosquitos_frame.merge(weather, how='left', on='Date')
```

```python
print(mosquitos_frame)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x='Tmin', y='NumMosquitos', data=mosquitos_frame)
plt.show()
```
